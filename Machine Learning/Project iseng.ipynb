```python
# Menampilkan DataFrame Internet Firewall Data 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
df = pd.read_csv(".\log2.csv")
df
```

```python
# Memeriksa missing value
for i in df.columns:
    miss = df[i].isnull().sum()
    if miss>0:
        print('\033[1m', '\33[31m')
        print("{} terdapat {} missing value(s)".format(i,miss))
    else:
        print('\033[1m', '\33[34m')
        print("{} tidak terdapat missing value!".format(i))
```

```python
# memisahkan kolom 'Action' dari dataframe df dan menyimpannya ke variabel y
# dalam bentuk matriks 2 dimensi, serta menyimpan dataframe df tanpa kolom 'Action' ke variabel x_data
y = df['Action'].values
y = y.reshape(-1,1)
x_data = df.drop(['Action'],axis = 1)
print(x_data)
```

```python
# membuat sebuah plot visualisasi jumlah pengamatan (count) pada setiap kategori yang terdapat pada kolom 'Action'
sns.countplot(x='Action',data=df,palette='hls')
plt.show()
```

### Normalisasi DataFrame

```python
import numpy as np
import pandas as pd

# normalisasi dataframe x_data
min_val = np.min(x_data)
max_val = np.max(x_data)
x = (x_data - min_val) / (max_val - min_val)

# tampilkan 5 baris teratas dari dataframe x
print(x.head())
```

### Klasifikasi

```python
from sklearn.model_selection import train_test_split

# membagi dataset menjadi data latih dan data uji
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=2500)

# menampilkan jumlah baris dan kolom dari setiap subset
print(f"Jumlah baris dan kolom dari x_train: {x_train.shape}")
print(f"Jumlah baris dan kolom dari x_test: {x_test.shape}")
print(f"Jumlah baris dan kolom dari y_train: {y_train.shape}")
print(f"Jumlah baris dan kolom dari y_test: {y_test.shape}")
```

### Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, plot_confusion_matrix
import matplotlib.pyplot as plt
from sklearn import tree

# inisialisasi objek DecisionTreeClassifier
dt = DecisionTreeClassifier()

# melatih model Decision Tree pada data latih
dt.fit(x_train, y_train)

# membuat prediksi menggunakan model yang telah dilatih pada data uji
y_forecast = dt.predict(x_test)

# menampilkan classification report
print(classification_report(y_test, y_forecast))

# menampilkan akurasi model
accuracy = accuracy_score(y_test, y_forecast)
print('\033[1m', '\33[32m')
print(f'Akurasi model Decision Tree: {accuracy:.4f}')

# menampilkan confusion matrix
plot_confusion_matrix(dt, x_test, y_test, cmap=plt.cm.Blues)

# menampilkan visualisasi model Decision Tree
plt.figure(figsize=(30,30))
tree.plot_tree(dt, fontsize=10)
print('\033[1m', '\33[32m')
plt.show()



```

### Klasifikasi Naive Bayes 

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, accuracy_score

# inisialisasi objek GaussianNB
nb = GaussianNB()

# melatih model Naive Bayes pada data latih
nb.fit(x_train, y_train.ravel())

# membuat prediksi menggunakan model yang telah dilatih pada data uji
y_forecast = nb.predict(x_test)

# menampilkan classification report
print(classification_report(y_test, y_forecast))

# menampilkan akurasi model
accuracy = accuracy_score(y_test, y_forecast)
print('\033[1m', '\33[32m')
print(f'Akurasi model Klasifikasi Naive Bayes : {accuracy:.4f}')
```

### Klasifikasi Regresi Logistik

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# inisialisasi objek LogisticRegression
lr = LogisticRegression(solver='lbfgs')

# melatih model Logistic Regression pada data latih
lr.fit(x_train, y_train.ravel())

# membuat prediksi menggunakan model yang telah dilatih pada data uji
y_forecast = lr.predict(x_test)

# menampilkan classification report
print(classification_report(y_test, y_forecast))

# menampilkan akurasi model
accuracy = accuracy_score(y_test, y_forecast)
print('\033[1m', '\33[32m')
print(f'Akurasi model: {accuracy:.4f}')
```

### K-Nearest Neighbors

```python
# Import library yang dibutuhkan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Inisialisasi model KNN Classifier dengan k = 1
K = 2
knn = KNeighborsClassifier(n_neighbors=K)
knn = knn.fit(x_train,y_train.ravel())

# Evaluasi performa model pada data pengujian dan data pelatihan
print(f"k = {K} neighbors, knn test: {knn.score(x_test, y_test)}")
print(f"knn = {K} neighbors, knn train: {knn.score(x_train, y_train)}")

# Mencari nilai k yang paling optimal
ran = np.arange(1, 40)
train_list = []
test_list = []
for each in ran:
    knn = KNeighborsClassifier(n_neighbors=each)
    knn = knn.fit(x_train, y_train.ravel())
    test_list.append(knn.score(x_test, y_test))
    train_list.append(knn.score(x_train, y_train))
    
best_test = np.max(test_list)
best_test_k = test_list.index(best_test) + 1
best_train = np.max(train_list)
best_train_k = train_list.index(best_train) + 1

print(f"best test: {best_test}, k = {best_test_k}")
print(f"best train: {best_train}, k = {best_train_k}")

# Plot performa model terhadap nilai k
plt.plot(ran, train_list, label='Train Score')
plt.plot(ran, test_list, label='Test Score')
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
plt.legend()
plt.show()
```
